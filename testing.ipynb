```python
def radarWidth(tif_path):
    # Given a file path to a .tif file, return the number of columns of the raster
    print(tif_path)
    radar = Image.open(tif_path)
    radar_mat = np.array(radar)
    length = radar_mat.shape[1]
    return length
```
